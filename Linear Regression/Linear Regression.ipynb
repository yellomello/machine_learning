```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_excel("worldincome.xlsx", skiprows=4)
```

```python
df.head(12)

```

```python
%matplotlib inline
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
d=pd.read_excel("areas.xlsx")
```

```python
d.head(17)
```

```python
p=reg.predict(d)
```

```python
d['predicted']=p
```

```python
d
```

```python
d.to_csv("prediction.csv")
```
