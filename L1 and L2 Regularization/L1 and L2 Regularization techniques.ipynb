```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Melbourne_housing_FULL.csv')
```

```python
df.head()
```

```python
cols_to_use=['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car','Landsize', 'BuildingArea', 'Price']
```

```python
df=df[cols_to_use]
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero=['Propertycount', 'Bedroom2', 'Distance', 'Bathroom', 
                   'Car']
df[cols_to_fill_zero]=df[cols_to_fill_zero].fillna(0)
```

```python
df
```

```python
df.isna().sum()
```

```python
df['Landsize']=df['Landsize'].fillna(df.Landsize.mean())
```

```python
df['BuildingArea']=df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
len(df)
```

```python
df.isna().sum()
```

```python
#we will need to get dummies for a lot of variables here
```

```python
df=pd.get_dummies(df, drop_first=True)
df.head()
```

```python
X=df.drop('Price', axis=1)
```

```python
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2, random_state=2)
```

```python
#we will use our linear regression algorithm here
from sklearn.linear_model import LinearRegression, Lasso , Ridge
reg=LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
lasso_reg=Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
ridge_reg=Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```
