```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("carprices.csv")
```

```python
df
```

```python
df.columns = df.columns.str.replace(' ','_')
dummies= pd.get_dummies(df.Car_Model)
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
```

```python
merged
```

```python
merged.columns = merged.columns.str.replace(' ','_')
final=merged.drop(['Car_Model', 'Audi_A5'], axis='columns') 
```

```python
final=final.rename(columns={'Sell_Price($)': 'sell'})
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
```

```python
 #indepedent Variables go in X
    
X= final.drop('sell', axis='columns')
X
```

```python
y=final.sell
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.score(X,y)
```
